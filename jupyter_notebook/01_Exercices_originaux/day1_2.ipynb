## Comparison Operators
Comparison operators take two variables and return a boolean.
- `<`: Smaller than.
- `>`: Greater than.
- `<=`: Smaller or equal than.
- `>=`: Greater or equal than.
- `==`: Equal to.
- `!=`: Unequal to.

## Logical Operators
Logical operators take two boolean variables and return a boolean.
- `and`
- `or`
- `not`

```python
maria_name = "maria"
maria_age = 5
maria_money = 1500.0
iphone_price = 899.9

is_maria_maria = maria_name == 'maria'

is_maria_5 = maria_age == 5

can_maria_drive = maria_age >= 18

can_maria_buy_iphone = maria_money >= iphone_price

# Equivalently
can_maria_buy_iphone = (maria_money - iphone_price) >= 0

# Equivalently
can_maria_buy_iphone = not((maria_money - iphone_price) < 0)

can_maria_drive_and_buy_iphone = can_maria_drive and can_maria_buy_iphone
```

### Dealing with Exceptions
Sometimes Exceptions must not stop code from running...
We use `try` and `except`.

```python
a = 3.0

# We assume b has a value unknown by the programmer, and thus it can be 0
b = input("Input a value for 'b', can be zero.")

try:
    c = a / float(b)
except ZeroDivisionError:
    c = 0
except:
    c = -1

print(c)
```

## Opening files
Use `open` command with file path and open mode, it can be:
- 'r'	Open a file for reading. (default)
- 'w'	Open a file for writing. Creates a new file if it does not exist or truncates the file if it exists.
- 'x'	Open a file for exclusive creation. If the file already exists, the operation fails.
- 'a'	Open for appending at the end of the file without truncating it. Creates a new file if it does not exist.
- 't'	Open in text mode. (default)
- 'b'	Open in binary mode.
- '+'	Open a file for updating (reading and writing)

## Branchings (aka Conditional statements)
- Single-Path: `if` block.
- Two-Path: `if-else` block.
- Multi-Path: `if-elif-...-elif-else` block.

```python
maria_age = 18

# We want to see if Maria can drive.
can_maria_drive = maria_age >= 18
if can_maria_drive:
    print("Maria can legally drive!")
else:
    print("Maria cannot get a driving license yet!")
```

```python
car_deposit_level_percent = 25.0

# We want to get a status message of our gas level.
if car_deposit_level_percent < 10:
    print("Find a gas station ASAP!")
elif car_deposit_level_percent < 30:
    print("The gas level is low, might need to refill.")
elif car_deposit_level_percent <= 100:
    print("You're doing great, keep driving!")
else:
    print("Gas level above maximum, check for leakage.")
```

```python
# EQUIVALENTLY:
car_deposit_level_percent = 25.0

if car_deposit_level_percent > 100:
    print("Gas level above maximum, check for leakage.")
else:
    if car_deposit_level_percent < 10:
        print("Find a gas station ASAP!")
    if car_deposit_level_percent < 30:
        print("The gas level is low, might need to refill.")
    else :
        print("You're doing great, keep driving!")
```

## Match Statement

```python
maria_age = 3

print("Happy bday Maria!")
match maria_age:
    case 21:
        print("Maria can now drink in the US!")
    case 18:
        print("Maria is now an adult! Careful for prison 💀")
    case 65:
        print("Maria can legally retire!")
    case 3:
        print("Maria can start Kids school.")
    case _:
        print("Nothin special happens on this age 😢")
```
