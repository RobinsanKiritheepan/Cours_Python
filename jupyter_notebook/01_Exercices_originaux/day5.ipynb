## Data Structures: Series and DataFrames

Pandas relies on two foundational data structures:
1.	**Series:** A **1-dimensional** labeled array (like a single column in an Excel sheet).
2.	**DataFrame:** A **2-dimensional** labeled data structure (like an entire Excel spreadsheet or SQL table).

```python
# The standard alias for pandas used across the globe is 'pd'
import pandas as pd

# 1. Creating a DataFrame from a dictionary of lists
employee_data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Department": ["HR", "Engineering", "Engineering", "Marketing"],
    "Salary": [65000, 85000, 92000, 72000],
    "Years_Exp": [3, 5, 7, 2]
}

df = pd.DataFrame(employee_data)

# 2. Inspecting the DataFrame
print("--- Displaying the DataFrame ---")
display(df)  # In Jupyter notebooks, display() renders interactive HTML tables!
```

## Getting Info about a DataFrame
- `shape`: Returns the shape (rows and columns) of a DataFrame.
- `head()`: Returns the first elements of a DataFrame.
- `tail()`: Returns the last elements of a DataFrame.
- `dtypes`: Returns the data types of each column.
- `info()`: Prints information about a DataFrame including the index dtype and columns, non-null values and memory usage.

```python
print("\n--- Shape and Quick Look ---")
print(f"Shape (Rows, Columns): {df.shape}")

print("\n--- Statistical Information ---")
display(df.describe())

print("\n--- First Elements ---")
display(df.head())

print("\n--- Last Elements ---")
display(df.tail())
```

```python
print("\n--- Summary of technical info ---")
df.info()
```

```python
print("\n--- Types of each Column ---")
print(df.dtypes)
```

## Accessing Columns

We access a column with the operator `[]`. E.g. `df["Gender"]`.\
To access multiple columns, we can put a list inside `[]`. E.g. `df[["Gender", "Blood type"]]`

```python
departments = df["Department"]
print(departments)
```

```python
name_with_department = df[["Name", "Department"]]
display(name_with_department)
```

## Accessing Rows: By Index

Each row has an underlying index. Similarly to a Python List.
We use `.iloc[index]`.

Slicing can be done with `.iloc[start:end]`.

```python
first_row = df.iloc[0]
display(first_row)
```

```python
first_two_rows = df.iloc[0:2]
display(first_two_rows)
#display(first_two_rows.iloc[1])
```

## Accessing Rows: By Value

```python
# We want to get the row whose "Name" is "Bob".

bobs_row = df[df["Name"] == "Bob"]
display(bobs_row)
```

## Slicing a DataFrame: Bidimensional Slicing

- `.iloc[row_slice, col_slice]`: Returns a view (DataFrame) sliced by rows and columns. It uses **indices** to slice.
- `.loc[row_slice, col_slice]`: Returns a view (DataFrame) sliced by rows and columns. It uses **labels** to slice.

```python
display(df.iloc[0:2, 0:2])
display(df.iloc[[0,1,2,3], [0,1,2,3]])
```

```python
display(df.loc[[0,1,2], ["Salary", "Name"]])
display(df.loc[0:2, ["Salary", "Name"]])
```

## Adding/Modifying a Column

```python
# Adding a new column (Bonus calculated as 10% of salary)
df["Bonus"] = df["Salary"] * 0.10

# Modifying an existing column (Giving everyone a $2,000 raise)
df["Salary"] = df["Salary"] + 2000

print("--- DataFrame after Modifications ---")
display(df)
```

## Basic Aggregations
Pandas provides built-in statistical functions to summarize numeric attributes instantly.
- `.mean()`: Calculates the average.
- `.sum()`: Totals up values.

```python
how_many_rows = len(df)
print(df[["Salary","Years_Exp"]].mean)

mean_salary = df['Salary'].mean()
print(f"Average Salary: ${mean_salary:.2f}")

total_bonus = df['Bonus'].sum()
print(f"Total Bonus Payout: ${total_bonus:.2f}")
```

## Grouping Data (Group By)
When working with tables, you often want to aggregate data by categories (e.g., finding the average salary per department). We use `df.groupby('Column_Name')` followed by an aggregation function.

```python
# Find the average salary by department
dept_salaries = df.groupby("Department")["Salary"].mean()

print("--- Average Salary by Department ---")
print(dept_salaries)
```

## Handling Missing Data (NaN)
Real-world data is rarely perfect. Pandas uses NaN (Not a Number) to represent empty cells.
- `.isna()`: Detects missing data.
- `.fillna(value)`: Replaces missing values with a designated default.
- `.dropna()`: Deletes any row containing missing values.

```python
import numpy as np

# Let's introduce some missing data for demonstration
corrupted_data = df.copy()
corrupted_data.loc[1, "Salary"] = np.nan

print("--- Data with Missing Salary ---")
display(corrupted_data)

# Fix it by filling NaN with the average salary of the remaining team
avg_salary = corrupted_data["Salary"].mean()
fixed_data = corrupted_data.fillna({"Salary": avg_salary})

print("\n--- Fixed Data (NaN filled with average) ---")
display(fixed_data)
```
