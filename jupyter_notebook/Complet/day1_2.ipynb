## 3. Les opérateurs arithmétiques

|---|---|---|---|
| `+` | addition | `3 + 2` | `5` |
| `-` | soustraction | `3 - 2` | `1` |
| `*` | multiplication | `3 * 2` | `6` |

**Priorité des opérations** : comme en maths — parenthèses d'abord, puis `**`, puis `* / // %`, puis `+ -`. En cas de doute : mets des parenthèses, c'est gratuit et lisible.

## 5. Les opérateurs logiques

Ils combinent des booléens entre eux :

| Opérateur | Résultat `True` si… | Table rapide |
|---|---|---|
| `and` (ET) | les DEUX conditions sont vraies | `True and True → True`, tout le reste → `False` |
| `or` (OU) | AU MOINS UNE condition est vraie | `False or False → False`, tout le reste → `True` |
| `not` (NON) | inverse la valeur | `not True → False` |

**⚠️ Priorité :** `not` passe avant `and`, qui passe avant `or`. Donc `A and B or C` = `(A and B) or C`. Quand tu mélanges `and` et `or`, mets TOUJOURS des parenthèses pour rendre ton intention explicite — c'est la source n°1 de bugs logiques.

````python
maria_name = "maria"
maria_age = 5
maria_money = 1500.0
iphone_price = 899.9

is_maria_maria = maria_name == 'maria'

is_maria_5 = maria_age == 5

can_maria_drive = maria_age >= 18

can_maria_buy_iphone = maria_money >= iphone_price

# Equivalently
can_maria_buy_iphone = (maria_money - iphone_price) >= 0

# Equivalently
can_maria_buy_iphone = not((maria_money - iphone_price) < 0)

can_maria_drive_and_buy_iphone = can_maria_drive and can_maria_buy_iphone
````

## 7. Les erreurs et les exceptions

Tout programme peut rencontrer 2 familles d'erreurs :

**1. Erreur de SYNTAXE (`SyntaxError`)** : Python ne COMPREND même pas ton code (faute de grammaire). Le programme ne démarre pas du tout. Ex : `a eq 2` — « eq » n'existe pas en Python.

**2. Erreur d'EXÉCUTION (*runtime*)** : le code est bien écrit, Python le lance… mais une ligne demande l'impossible. Python lève alors une **Exception** et le programme s'arrête net. Les 4 grandes classiques :

| Exception | Cause | Exemple |
|---|---|---|
| `TypeError` | opération entre types incompatibles | `3 + "hello"` (nombre + texte) |
| `NameError` | variable jamais définie (souvent une faute de frappe !) | écrire `gap` au lieu de `gpa` |
| `ZeroDivisionError` | division par zéro | `3.8 / 0` |
| `ValueError` | bon type, mais valeur invalide | `int("abc")` |

**💡 Lire un message d'erreur (compétence n°1 du programmeur !) :** lis-le de BAS en HAUT — la dernière ligne donne le type d'exception et l'explication ; juste au-dessus, la flèche montre la ligne fautive. ### Gérer les exceptions : `try` / `except`

Parfois une erreur est **prévisible** (ex : l'utilisateur peut taper n'importe quoi) et ne doit PAS faire planter le programme. Structure :

```python
try:
    # le code « risqué »
except ZeroDivisionError:
    # exécuté SEULEMENT si cette erreur précise survient
except:
    # exécuté pour TOUTE autre erreur (filet de sécurité)
```

**Le déroulement :** Python essaie le bloc `try` ; à la première erreur, il SAUTE directement dans le `except` correspondant (le reste du `try` est abandonné) ; s'il n'y a aucune erreur, les `except` sont ignorés.

**⭐ Bonne pratique :** attraper des exceptions PRÉCISES (`except ValueError:`) plutôt qu'un `except:` nu qui masque tout, y compris tes vraies fautes de code.

````python
a = 3.0

# We assume b has a value unknown by the programmer, and thus it can be 0
b = input("Input a value for 'b', can be zero.")

try:
    c = a / float(b)
except ZeroDivisionError:
    c = 0
except:
    c = -1

print(c)
````

## 8. La bibliothèque `time`

Deux outils à connaître :
- **`time.perf_counter()`** : un chronomètre de haute précision. On le lit avant et après un bout de code, et la différence = le temps d'exécution. C'est l'outil standard pour mesurer les performances.
- **`time.sleep(n)`** : met le programme en pause pendant `n` secondes.

**Le pattern chronomètre à mémoriser :**
```python
start = time.perf_counter()
# ... code à mesurer ...
end = time.perf_counter()
print(end - start, "secondes")
```

````python
import time

start = time.perf_counter()
a = 3 + 2
time.sleep(2.0)

end = time.perf_counter()

print("Two seconds have passed.")
print(end-start, "seconds")
````

## 9. La bibliothèque `datetime`

Pour manipuler des dates et des heures :
- **`datetime.datetime.now()`** → la date et l'heure actuelles.
- **`datetime.datetime.strptime(texte, format)`** → convertit du TEXTE en objet date (*string parse time*). Il faut décrire le format du texte avec des codes :

| Code | Signification | Exemple |
|---|---|---|
| `%Y` | année sur 4 chiffres | 2026 |
| `%m` / `%d` | mois / jour | 07 / 09 |
| `%H` / `%M` | heures / minutes | 7 / 30 |

Ici `"2026, 7:30"` est décodé avec `"%Y, %H:%M"` — le format doit correspondre EXACTEMENT au texte (virgule et espaces compris !).

**Le bonus magique :** soustraire deux dates (`a - date`) donne un objet `timedelta` — la durée entre les deux, affichée en jours + heures:minutes:secondes.

````python
import datetime

a = datetime.datetime.now()

s = "2026, 7:30"

date = datetime.datetime.strptime(s, "%Y, %H:%M")

diff = a-date

print(diff, type(diff))
````

## 10. Les bibliothèques `subprocess`, `os` et `signal`

## Les branchements (structures conditionnelles)

C'est LE mécanisme qui permet au programme de **prendre des décisions** :
- **1 chemin** : `if` seul — « fais ceci SI la condition est vraie (sinon ne fais rien) »
- **2 chemins** : `if` / `else` — « ceci OU cela »
- **N chemins** : `if` / `elif` / … / `else` — une cascade de conditions testées DANS L'ORDRE ; la PREMIÈRE vraie gagne et les autres sont ignorées.

**⚠️ La syntaxe Python :** deux-points `:` en fin de ligne de condition, et le bloc en dessous **indenté** (décalé de 4 espaces). L'indentation n'est pas décorative en Python — c'est ELLE qui délimite le bloc !

### Démonstrations des branchements (3 cellules suivantes)
1. le `if`/`else` simple, 2. la cascade `if`/`elif`/`else`, 3. une version « équivalente » imbriquée… qui contient un bug subtil à repérer !

**Démo 1 — `if`/`else` :** `maria_age >= 18` donne un booléen ; le `if` choisit la branche. Avec 18 exactement → `>=` est vrai → « can legally drive ». Change la valeur et re-exécute pour voir l'autre branche !

````python
maria_age = 18

# We want to see if Maria can drive.
can_maria_drive = maria_age >= 18
if can_maria_drive:
    print("Maria can legally drive!")
else:
    print("Maria cannot get a driving license yet!")
````

**Démo 2 — la cascade `if`/`elif`/`else` :** avec 25.0 → le 1er test (< 10) est faux, le 2e (< 30) est VRAI → « gas level is low » et la cascade S'ARRÊTE là (le `<= 100`, pourtant vrai aussi, n'est jamais testé !). **C'est la propriété clé : dans une cascade, un seul bloc s'exécute — le premier vrai.** L'ordre des conditions compte donc énormément.

````python
car_deposit_level_percent = 25.0

# We want to get a status message of our gas level.
if car_deposit_level_percent < 10:
    print("Find a gas station ASAP!")
elif car_deposit_level_percent < 30:
    print("The gas level is low, might need to refill.")
elif car_deposit_level_percent <= 100:
    print("You're doing great, keep driving!")
else:
    print("Gas level above maximum, check for leakage.")
````

### 🔴 Version « équivalente »… avec un BUG — sauras-tu le voir ?

Cette réécriture imbriquée n'est PAS équivalente : dans le bloc `else`, il y a `if (< 10):` puis `if (< 30):` — deux `if` **INDÉPENDANTS** au lieu d'un `if`/`elif` chaîné ! Conséquence : avec un niveau de 5 %, la version originale affiche seulement « Find a gas station ASAP! », mais celle-ci affiche « ASAP! » **ET** « gas level is low » (5 < 10 ET 5 < 30, les deux ifs se déclenchent).

**⭐ Leçon :** `elif` ≠ deux `if` séparés. `elif` = « teste SEULEMENT si les précédents étaient faux ». Deux `if` = tests indépendants qui peuvent TOUS se déclencher.

````python
# EQUIVALENTLY:
car_deposit_level_percent = 25.0

if car_deposit_level_percent > 100:
    print("Gas level above maximum, check for leakage.")
else:
    if car_deposit_level_percent < 10:
        print("Find a gas station ASAP!")
    if car_deposit_level_percent < 30:
        print("The gas level is low, might need to refill.")
    else :
        print("You're doing great, keep driving!")
````

## L'instruction `match` (Python ≥ 3.10)

Le `match` compare UNE valeur à plusieurs motifs (`case`) — plus lisible qu'une longue cascade de `elif` quand on teste des valeurs précises :
```python
match age:
    case 18: ...    # si age == 18
    case 65: ...    # si age == 65
    case _:  ...    # le « joker » : tous les autres cas (comme else)
```
Le `case _` final attrape tout ce qui n'a pas matché avant. Seul le PREMIER case qui correspond s'exécute.

**Démo `match` :** avec `maria_age = 3` → le `case 3:` matche → « Kids school ». Essaie 18, 99 (→ joker `_`), etc.

````python
maria_age = 3

print("Happy bday Maria!")
match maria_age:
    case 21:
        print("Maria can now drink in the US!")
    case 18:
        print("Maria is now an adult! Careful for prison 💀")
    case 65:
        print("Maria can legally retire!")
    case 3:
        print("Maria can start Kids school.")
    case _:
        print("Nothin special happens on this age 😢")
````

## L'affectation conditionnelle (opérateur ternaire)

Une valeur qui dépend d'une condition, en UNE ligne :
```python
msg = "Oui" if age >= 18 else "Non"
```
Se lit : « msg vaut "Oui" si la condition est vraie, sinon "Non" ». Équivalent à 4 lignes de `if`/`else` — parfait pour les petits choix simples ; pour une logique complexe, préfère le `if` classique (lisibilité d'abord !).

**Démo ternaire :** 17 ≥ 18 est faux → `msg = "No"`. Une ligne au lieu de quatre.

````python
maria_age = 17

can_maria_drive_msg = "Yes" if maria_age >= 18 else "No"

print(can_maria_drive_msg)
````

## Les f-strings : le formatage moderne

Préfixe `f` devant la chaîne → les `{variables}` sont remplacées par leur valeur :
```python
f"{nom} a {age} ans"                # insertion simple
f"{gpa:.3f}"                        # format : 3 décimales
f"{gpa=}"                           # debug : affiche « gpa=0.2 » (nom ET valeur !)
```
Le `:.3f` = « float à 3 décimales ». Le `{var=}` est génial pour déboguer vite. Les f-strings sont LA méthode moderne — oublie la concaténation à coups de `+` et `str()`.

**Démo f-strings :** `{gpa=}` affiche `gpa=0.2` (pratique en debug !), et `{gpa:.3f}` affiche `0.200` (3 décimales). Remarque : `.3f` ARRONDIT à l'affichage, il ne modifie pas la variable.

````python
person_name = "maria"
age = 14
gpa = 0.2

print(f"{gpa=}")
msg = f"{person_name} is {age} years old and has a GPA of {gpa:.3f}."
print(msg)
````
