## 1. Les types de base

Python manipule 4 types fondamentaux :

| Type | Nom complet | Exemple | Description |
|---|---|---|---|
| `int` | *integer* (entier) | `3`, `-42`, `1000000` | nombre SANS virgule |
| `float` | *floating point* (flottant) | `2.1`, `-0.5`, `3.0` | nombre À virgule (attention : `3.0` est un float, pas un int !) |

| `str` | *string* (chaîne) | `"hello"`, `'a'` | du texte, entre guillemets simples ou doubles |

## 2. Déclarer une variable

Une **variable** est une boîte avec un nom, dans laquelle on range une valeur :
```python
a = 3
```
- à GAUCHE du `=` : le **nom** de la boîte (tu le choisis),
- à DROITE : la **valeur** qu'on y range,
- le `=` est l'**opérateur d'affectation** (à ne pas confondre avec `==` qui COMPARE !).

Python devine tout seul le type : `a = 3` → int ; `a = 3.0` → float ; `a = "3"` → str (du texte, même si ça ressemble à un nombre !).

### 🔍 Démonstration — lis le résultat ligne par ligne
- `a = 3` puis `print(type(a))` → `<class 'int'>` : Python a rangé un entier.
- `b = 2.1` → `<class 'float'>` : la virgule (le point en anglais !) fait basculer en flottant.
- `c = "hello"` → `<class 'str'>` : les guillemets font du texte.

**⭐ Réflexe à prendre :** en cas de doute sur une variable, `print(type(...))` te dit toujours à quoi tu as affaire.

````python
# a will get created with the value of 3, therefore, a is an integer.
a = 3
print(a) 
print(type(a))

b = 2.1
print(b)
print(type(b))

c = "hello"
print(c)
print(type(c))
````

## 3. Les opérateurs arithmétiques

| Opérateur | Nom | Exemple | Résultat |
|---|---|---|---|
| `+` | addition | `3 + 2` | `5` |
| `-` | soustraction | `3 - 2` | `1` |
| `*` | multiplication | `3 * 2` | `6` |
| `/` | division « vraie » | `7 / 2` | `3.5` — donne TOUJOURS un float, même `6/2` → `3.0` ! |
| `//` | division **entière** (plancher) | `7 // 2` | `3` — le quotient sans les décimales |
| `%` | **modulo** (reste) | `7 % 2` | `1` — le RESTE de la division |
| `**` | puissance | `2 ** 3` | `8` (2×2×2) |

**⭐ Le duo à retenir absolument : `//` et `%`.** Ils répondent ensemble à « combien de fois X rentre dans Y, et combien il reste ? » :
- `17 // 5` → `3` (5 rentre 3 fois dans 17)
- `17 % 5` → `2` (et il reste 2) — vérification : 3×5+2 = 17 ✓

**Priorité des opérations** : comme en maths — parenthèses d'abord, puis `**`, puis `* / // %`, puis `+ -`. En cas de doute : mets des parenthèses, c'est gratuit et lisible.

### 🔍 Démonstration — prédis chaque valeur AVANT de lire le résultat !
- `a = 3 + 2` → **5** (la nouvelle valeur ÉCRASE l'ancienne `a = 3`)
- `b = a + b` → 5 + 2 = **7** (on peut utiliser `b` des deux côtés du `=` : la droite est calculée d'abord)
- `c = a ** b` → 5⁷ = **78125**
- `d = 5.3 // 2` → **2.0** ⚠️ division entière, MAIS résultat float car 5.3 est un float !
- `f = int(5.6)` → **5** : `int()` TRONQUE (coupe les décimales), il n'arrondit pas !
- `g = 9 / 3 * (1 + 2)` → 9 ou 1 ? Réponse : **9.0**. Les parenthèses d'abord (3), puis `/` et `*` ont la MÊME priorité → on lit de gauche à droite : 9/3 = 3.0, puis 3.0×3 = 9.0.

````python
a = 3           # a = 3
a = 3 + 2       # a = 5
print("a = ",a)
b = 2           # b = 2
b = a + b       # b = 7
print("b = ",b)
c = a ** b      # c = ?
print("c = ",c)
d = 5.3 // 2    # d = ?
print("d = ",d)
e = b * 5       # e = 35
print("e = ",e)
f = int(5.6)    # f = 5
print("f = ",f)
g = 9 / 3 * (1 + 2)  # 9 or 1 ?
print("g = ",g)
````

## 5. Les opérateurs logiques

Ils combinent des booléens entre eux :

| Opérateur | Résultat `True` si… | Table rapide |
|---|---|---|
| `and` (ET) | les DEUX conditions sont vraies | `True and True → True`, tout le reste → `False` |
| `or` (OU) | AU MOINS UNE condition est vraie | `False or False → False`, tout le reste → `True` |
| `not` (NON) | inverse la valeur | `not True → False` |

**⚠️ Priorité :** `not` passe avant `and`, qui passe avant `or`. Donc `A and B or C` = `(A and B) or C`. Quand tu mélanges `and` et `or`, mets TOUJOURS des parenthèses pour rendre ton intention explicite — c'est la source n°1 de bugs logiques.

### 🔍 Démonstration — la logique avec Maria
Chaque ligne construit un booléen à partir de comparaisons :
- `maria_age >= 18` → `False` (elle a 5 ans) — le résultat de la comparaison est directement rangé dans la variable `can_maria_drive`.
- Remarque les **trois écritures équivalentes** de `can_maria_buy_iphone` : une même logique peut s'écrire de plusieurs façons — préfère TOUJOURS la plus lisible (la première ici).
- Dernière ligne : `and` combine deux booléens déjà calculés — c'est plus lisible que de tout écrire en une ligne géante.

````python
maria_name = "maria"
maria_age = 5
maria_money = 1500.0
iphone_price = 899.9

is_maria_maria = maria_name == 'maria'

is_maria_5 = maria_age == 5

can_maria_drive = maria_age >= 18

can_maria_buy_iphone = maria_money >= iphone_price

# Equivalently
can_maria_buy_iphone = (maria_money - iphone_price) >= 0

# Equivalently
can_maria_buy_iphone = not((maria_money - iphone_price) < 0)

can_maria_drive_and_buy_iphone = can_maria_drive and can_maria_buy_iphone
````

## 7. Les erreurs et les exceptions

Tout programme peut rencontrer 2 familles d'erreurs :

**1. Erreur de SYNTAXE (`SyntaxError`)** : Python ne COMPREND même pas ton code (faute de grammaire). Le programme ne démarre pas du tout. Ex : `a eq 2` — « eq » n'existe pas en Python.

**2. Erreur d'EXÉCUTION (*runtime*)** : le code est bien écrit, Python le lance… mais une ligne demande l'impossible. Python lève alors une **Exception** et le programme s'arrête net. Les 4 grandes classiques :

| Exception | Cause | Exemple |
|---|---|---|
| `TypeError` | opération entre types incompatibles | `3 + "hello"` (nombre + texte) |
| `NameError` | variable jamais définie (souvent une faute de frappe !) | écrire `gap` au lieu de `gpa` |
| `ZeroDivisionError` | division par zéro | `3.8 / 0` |
| `ValueError` | bon type, mais valeur invalide | `int("abc")` |

**💡 Lire un message d'erreur (compétence n°1 du programmeur !) :** lis-le de BAS en HAUT — la dernière ligne donne le type d'exception et l'explication ; juste au-dessus, la flèche montre la ligne fautive. Les cellules suivantes provoquent volontairement chaque erreur pour que tu apprennes à les reconnaître.

**Démo 1 — `SyntaxError` :** `a eq 2` n'est pas du Python (on écrit `a == 2` pour comparer). Remarque que Python refuse de démarrer : AUCUNE ligne n'est exécutée, même les correctes.

````python
# Syntax Error
a eq 2
a plus 3
````

**Démo 2 — `TypeError` :** `3 + "hello"` — additionner un nombre et du texte n'a pas de sens pour Python. (Curiosité : `"a" + "b"` marche — le `+` CONCATÈNE les textes — et `3 * "ab"` donne `"ababab"` !)

````python
# Runtime: Type Error
a = 3 + "hello"
````

**Démo 3 — `NameError` :** on définit `gpa` mais on affiche `gap` — variable inexistante. C'est presque toujours une **faute de frappe**. Note le commentaire de l'enseignant : Python ne vérifie ce genre d'erreur QU'AU moment d'exécuter la ligne (contrairement à d'autres langages qui vérifient avant de lancer).

````python
# Runtime: Name Error
gpa = 3.8
print(gap)

# PS: Crazy that this code is even executed! Sanity checks before running are very simple...
````

**Démo 4 — `ZeroDivisionError` :** diviser par zéro est mathématiquement impossible → exception. C'est typiquement l'erreur qu'on gérera avec `try/except` juste après (car un dénominateur peut venir d'une saisie utilisateur imprévisible).

````python
# Runtime: Zero Division Error
gpa = 3.8 / 0
````

### Gérer les exceptions : `try` / `except`

Parfois une erreur est **prévisible** (ex : l'utilisateur peut taper n'importe quoi) et ne doit PAS faire planter le programme. Structure :

```python
try:
    # le code « risqué »
except ZeroDivisionError:
    # exécuté SEULEMENT si cette erreur précise survient
except:
    # exécuté pour TOUTE autre erreur (filet de sécurité)
```

**Le déroulement :** Python essaie le bloc `try` ; à la première erreur, il SAUTE directement dans le `except` correspondant (le reste du `try` est abandonné) ; s'il n'y a aucune erreur, les `except` sont ignorés.

**⭐ Bonne pratique :** attraper des exceptions PRÉCISES (`except ValueError:`) plutôt qu'un `except:` nu qui masque tout, y compris tes vraies fautes de code.

### 🔍 Démonstration `try`/`except`, pas à pas
1. `input(...)` demande une saisie à l'utilisateur — ⚠️ elle renvoie TOUJOURS du texte (`str`), même si tu tapes `5` !
2. D'où le `float(b)` : conversion texte → nombre. Deux échecs possibles :
   - l'utilisateur tape `0` → `float("0")` marche, mais `a / 0.0` → `ZeroDivisionError` → attrapé par le 1er except → `c = 0` ;
   - l'utilisateur tape `abc` → `float("abc")` → `ValueError` → attrapé par le `except:` générique → `c = -1`.
3. Le résultat affiché ici (−1) montre que l'utilisateur avait tapé quelque chose de non convertible.

Le programme ne plante JAMAIS : chaque scénario d'erreur a sa sortie propre.

````python
a = 3.0

# We assume b has a value unknown by the programmer, and thus it can be 0
b = input("Input a value for 'b', can be zero.")

try:
    c = a / float(b)
except ZeroDivisionError:
    c = 0
except:
    c = -1

print(c)
````

## 8. La bibliothèque `time`

Deux outils à connaître :
- **`time.perf_counter()`** : un chronomètre de haute précision. On le lit avant et après un bout de code, et la différence = le temps d'exécution. C'est l'outil standard pour mesurer les performances.
- **`time.sleep(n)`** : met le programme en pause pendant `n` secondes.

**Le pattern chronomètre à mémoriser :**
```python
start = time.perf_counter()
# ... code à mesurer ...
end = time.perf_counter()
print(end - start, "secondes")
```

### 🔍 Démonstration du chronomètre
`start` → on fait un petit calcul + `sleep(3.0)` → `end`. Résultat : ≈ 3.0004 s — les 3 secondes de pause + une fraction de milliseconde pour le calcul. (Petit décalage entre le texte affiché « Two seconds » et le `sleep(3.0)` — les prints mentent parfois, les mesures non 😄.)

````python
import time

start = time.perf_counter()
a = 3+2
time.sleep(3.0)

end = time.perf_counter()

print("Two seconds have passed")
print(end - start,"seconds")
````

## 9. La bibliothèque `datetime`

Pour manipuler des dates et des heures :
- **`datetime.datetime.now()`** → la date et l'heure actuelles.
- **`datetime.datetime.strptime(texte, format)`** → convertit du TEXTE en objet date (*string parse time*). Il faut décrire le format du texte avec des codes :

| Code | Signification | Exemple |
|---|---|---|
| `%Y` | année sur 4 chiffres | 2026 |
| `%m` / `%d` | mois / jour | 07 / 09 |
| `%H` / `%M` | heures / minutes | 7 / 30 |

Ici `"2026, 7:30"` est décodé avec `"%Y, %H:%M"` — le format doit correspondre EXACTEMENT au texte (virgule et espaces compris !).

**Le bonus magique :** soustraire deux dates (`a - date`) donne un objet `timedelta` — la durée entre les deux, affichée en jours + heures:minutes:secondes.

### 🔍 Démonstration datetime
`strptime("2026, 7:30", "%Y, %H:%M")` fabrique la date « 1er janvier 2026, 7h30 » (jour et mois non précisés → 01/01 par défaut). `a - date` → le temps écoulé depuis, en jours + h:min:s. Utile pour calculer des âges, des délais, des durées entre événements.

````python
import datetime

a = datetime.datetime.now()

s = "2026, 7:30"

date = datetime.datetime.strptime(s,"%Y, %H:%M")



print(a-date)
````

## 10. Les bibliothèques `subprocess`, `os` et `signal`

Pour piloter le SYSTÈME depuis Python :
- **`subprocess.Popen(...)`** : lance un **nouveau processus** (un autre programme) — ici la commande `dir` du terminal Windows. `shell=True` = passer par l'interpréteur de commandes.
- **`new_process.pid`** : le numéro d'identification du processus (*Process ID*) — chaque programme qui tourne sur ta machine en a un (visible dans le Gestionnaire des tâches).
- **`os.kill(pid, signal)`** : envoie un **signal** à un processus — ici `SIGINT` (signal d'interruption, l'équivalent de Ctrl+C).

⚠️ Section « culture générale » : tu utiliseras rarement ça en data science, mais c'est la base de l'automatisation système.

### 🔍 Démonstration subprocess
On lance `dir hello` dans un sous-processus, on récupère son `pid`, et on lui envoie `SIGINT`. Le `print` montre l'objet `Popen` avec `returncode: None` (le processus n'avait pas encore fini au moment du print). Ne t'inquiète pas si cette cellule se comporte différemment chez toi — la gestion des signaux varie selon Windows/Mac/Linux.

````python
import os 
import subprocess
import signal

#Create new process
new_process = subprocess.Popen("dir hello", shell=True)

#Send a signal to a process, given its PID.
os.kill(new_process.pid, signal.SIGINT)

print(new_process)
````

## 11. Ouvrir des fichiers

**Méthode basique :** `open("mon.txt")` ouvre le fichier (en lecture par défaut) ; `next(file)` lit la ligne suivante ; `file.close()` le referme.

**🔴 BUG dans la cellule ci-dessous — cherche-le :** il est écrit `file.close` SANS parenthèses ! Sans les `()`, on ne fait que MENTIONNER la fonction (Python affiche `<function ...close()>`), on ne l'APPELLE pas — le fichier n'est jamais vraiment fermé. C'est exactement pour éviter ce genre d'oubli que le `with` (cellule suivante) existe.

````python
file = open("mon.txt")
print(next(file))
file.close
````

````python

# With context manager, open and close is handled automatically


# Opens as read only
with open("mon.txt") as file:
    print(next(file))


# Opens with writting permission, removes old content
with open("mon.txt", "w") as file:
    file.write("Hello mom\n")


# Opens with writting permission, with old content
with open("mon.txt", "a") as file:
    file.write("Hello dad")
````
