## Dictionaries

Built-in data type used to store data in key-value pairs.

Dictionary in Python is an unordered, mutable collection of key-value pairs. Think of it like a real-world dictionary: you look up a "word" (the key) to find its "definition" (the value).
Key Rules of Dictionaries:
- Keys must be **unique** and **immutable** (strings, numbers, or tuples).
- Values can be **anything** (lists, strings, integers, or even other dictionaries).
- Syntactically, they use curly braces `{}` and colons `:` to separate keys and values.

```python
# Creating a dictionary to represent a spaceship's status
spaceship = {
    "name": "Apollo Python",
    "crew_count": 3,
    "fuel_level": 84.5,
    "cargo": [" freeze-dried ice cream", "oxygen tanks", "space wrenches"],
    "shield_active": True
}

# 1. Accessing data using keys
print(f"🚀 Spaceship Name: {spaceship['name']}")
print(f"🎒 Cargo Item 1: {spaceship['cargo'][0]}")

# 2. Modifying and Adding new data
spaceship["fuel_level"] = 79.2  # Updating an existing key
spaceship["destination"] = "Mars"  # Adding a completely new key-value pair

# 3. Checking if a key exists safely
if "shields_active" in spaceship:
    print(f"Shield Status: {spaceship['shields_active']}")
else:
    print("⚠️ 'shields_active' key not found! Did you misspell it?")

# 4. Looping through a dictionary
print("\n--- Full Spaceship Diagnostic Report ---")
for key, value in spaceship.items():
    print(f"System [{key.upper()}]: {value}")
```

## Useful Dictionary Methods
- `.get(key, default)`: Safely looks up a key. If the key doesn't exist, it returns a default value (like None or a message) instead of crashing your program.
- `.pop(key)`: Removes a key and returns its value so you can use it or store it in a variable.
- `.keys()`: Returns a view of all the keys in the dictionary.
- `.values()`: Returns a view of all the values in the dictionary.
- `.update(other_dict)`: Merges another dictionary into the current one, overwriting any matching keys.
- `.items()`: Returns an iterable with all key-value pairs as tuples. Can be used to either iterate the dictionary or modify its content.


```python
# A dictionary representing a smart home system configuration
smart_home = {
    "living_room_temp": 22,
    "lights_on": ["kitchen", "hallway"],
    "security_armed": True,
    "eco_mode": False
}

# 1. Safe Access with .get()
# Standard bracket lookups crash if the key is missing. .get() handles it gracefully.
thermostat = smart_home.get("backyard_temp", "Sensor Offline")
print(f"🌡️ Backyard Temperature: {thermostat}")

# 2. Extracting Keys and Values independently
print(f"🔑 Available Settings: {list(smart_home.keys())}")
print(f"📊 Current Values: {list(smart_home.values())}")

# 3. Removing items safely using .pop()
# Let's turn off eco mode and capture its last state
was_eco = smart_home.pop("eco_mode")
print(f"♻️ Removed 'eco_mode'. Its last setting was: {was_eco}")

# 4. Merging dictionaries with .update()
# Imagine the user applies a 'Night Mode' preset
night_preset = {
    "living_room_temp": 19,
    "lights_on": ["porch"],
    "camera_recording": True # This is a brand new setting!
}

smart_home.update(night_preset)

print("\n--- Updated Smart Home Status (Night Mode) ---")
for setting, status in smart_home.items():
    print(f"{setting}: {status}")
```

## Variable Positional Arguments (*args)
How do we create a function that **we do not know how many parameters** will receive?

For example, the `print` function. It can get any number of arguments, and will print all of them.

```python
def plan_trip(desination_country: str, *cities):
    print(f"Trip planner to {desination_country.upper()}")
    print(f"The type of 'cities' is {type(cities)}")
    print(f"Content of 'cities': {cities}")

    # Because 'cities' is a tuple, we can check its length and loop through it!
    if len(cities) == 0:
        print("Warning: No cities added to this itinerary yet.")
    else:
        print("Stops along the way:")
        for city in cities:
            print(f"   {city}")

plan_trip("Japan", "Tokyo", "Kyoto", "Osaka")
```

**EXERCISE:** Write a function called `multiply_all(*numbers)` that takes any number of numeric arguments and returns their product (all of them multiplied together).

```python
def multiply_all(*numbers):
    if len(numbers) == 0:
        return 0
    product = 1
    for n in numbers:
        product *= n
    return product
    
    
    
print(multiply_all(2, 3, 4))        # Expected Output: 24
print(multiply_all(5, 10, -5, 9))   # Expected Output: -2250
print(multiply_all())               # Expected Output: 0
```

## Variable Keyword Arguments (**kwargs)
`**kwargs` allows a function to accept any number of named (keyword) arguments.

Inside the function, Python packs these arguments into a Dictionary!

```python
def build_profile(first_name, last_name, **additional_info):
    """Builds a dictionary containing everything we know about a user."""
    profile = {
        "first": first_name,
        "last": last_name,
    }
    
    # additional_info is a dictionary containing any extra keyword arguments passed!
    print(f"The type of additional_info is {type(additional_info)}")
    print(f"Content of **kwargs: {additional_info}\n")
    
    for key, value in additional_info.items():
        profile[key] = value
        
    return profile

# Scenario A: Minimal info
user1 = build_profile("Ada", "Lovelace")
print("User 1 Profile:", user1)
print("-" * 20)

# Scenario B: Lots of optional info
user2 = build_profile("Albert", "Einstein", 
                      location="Princeton", 
                      field="Physics", 
                      nobel_prize=True)
print("User 2 Profile:", user2)
```

## Object-Oriented Programming (OOP)

So far, we have used Procedural Programming: writing code as a sequence of steps, functions, and logic blocks.

Object-Oriented Programming (OOP) is a different paradigm. It allows us to group data (attributes) and behavior (methods) together into a single blueprint called a Class. An individual object created from that blueprint is called an Instance.
- Class: The blueprint/template (e.g., The architectural blueprint for a house).
- Object/Instance: The actual physical thing built from the blueprint (e.g., The physical house at 123 Main Street).
- Attributes: Variables inside a class (what the object knows about itself).
- Methods: Functions inside a class (what the object can do).

```python
class VirtualPet:
    # The Constructor method - runs automatically when a new pet is created
    def __init__(self, name, species):
        self.name = name          # Attribute
        self.species = species    # Attribute
        self.energy = 100         # Default Attribute value

    # A Method representing behavior
    def play(self, minutes):
        energy_lost = minutes * 2
        self.energy -= energy_lost
        print(f"🎮 {self.name} played for {minutes} minutes and lost {energy_lost} energy.")
        if self.energy < 20:
            print(f"😴 {self.name} is exhausted! Time for a nap.")

    # A Method to check status
    def check_status(self):
        print(f"📊 Pet Status -> Name: {self.name} | Species: {self.species} | Energy: {self.energy}%")


# --- Putting the Blueprint to Work (Instantiating Objects) ---

# Creating two distinct pets from the same blueprint
pet1 = VirtualPet("Sparky", "Dog")
pet2 = VirtualPet("Whiskers", "Cat")

# Interacting with our objects
pet1.check_status()
pet1.play(20)
pet1.check_status()

print("-" * 30)

pet2.check_status()
pet2.play(45)
```

## Inheritance
Imagine you are building a video game. You have a Warrior class, a Mage class, and a Rogue class. They all have a name, health, and a take_damage() method. Writing that exact same code three separate times is a violation of the programming principle DRY (Don't Repeat Yourself).

Inheritance allows us to create a generic Parent Class (e.g., Character) and then create specific Child Classes (e.g., Warrior) that automatically inherit all attributes and methods from the parent.
- Method Overriding: A child class can completely redefine a method inherited from its parent to change its behavior.

```python
# The Parent Class (Generic Blueprint)
class Media:
    def __init__(self, title, duration_mins):
        self.title = title
        self.duration_mins = duration_mins
        self.views = 0

    def play(self):
        self.views += 1
        print(f"▶️ Playing: '{self.title}'")

    # Special method to handle how this object prints as a string
    def __str__(self):
        return f"'{self.title}' ({self.duration_mins} mins) - Views: {self.views}"


# The Child Class (Inherits everything from Media)
class Movie(Media):
    # Overriding the parent constructor to add a specific attribute
    def __init__(self, title, duration_mins, director):
        # super().__init__ calls the parent class's constructor to set title and duration
        super().__init__(title, duration_mins)
        self.director = director  # Specific to Movies

    # Overriding the parent's print string representation
    def __str__(self):
        return f"🎬 Movie: '{self.title}' directed by {self.director}"


# Another Child Class
class Series(Media):
    def __init__(self, title, duration_mins, total_episodes):
        super().__init__(title, duration_mins)
        self.total_episodes = total_episodes

    # Overriding the play method specifically for Series behavior
    def play(self):
        self.views += 1
        print(f"📺 Binging an episode of '{self.title}' ({self.total_episodes} eps total)")


# --- Testing Inheritance & Special Methods ---

# 1. Look how clean the print statements look now because of __str__
generic_video = Media("Funny Cat Compilation", 5)
print(generic_video) 

movie1 = Movie("Inception", 148, "Christopher Nolan")
print(movie1) 

series1 = Series("Python for Beginners Show", 45, 10)
print(series1)

print("-" * 40)

# 2. Testing inherited and overridden methods
movie1.play()       # Uses parent's play() method logic
series1.play()      # Uses its own overridden play() method logic

print(f"Movie Views: {movie1.views}")
print(f"Series Views: {series1.views}")
```

## Inspecting Objects
- `isinstance(obj, Class)` allows to check if an object belongs to a specific class.
- `hasattr(Class, "attribute")` checks if a class has an attribute defined.
- `issubclass(obj, Class)` checks if an object belongs to a subclass of a class.
- `issuperclass(obj, Class)` checks if an object belongs to a superclass of a class.

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def bark(self):
        return "Woof!"

# Let's create a dog object
my_dog = Dog("Buddy")

# 1. Checking Class Relationships with issubclass()
print("--- Class Checks ---")
print(issubclass(Dog, Animal))     # True (Dog is a child of Animal)
print(issubclass(Animal, Dog))     # False (Animal is not a child of Dog)

# 2. Checking Objects with isinstance()
print("\n--- Object Checks ---")
print(isinstance(my_dog, Dog))     # True (It is a Dog)
print(isinstance(my_dog, Animal))  # True (It is also an Animal because of inheritance!)

# 3. Checking Features with hasattr()
print("\n--- Feature Checks ---")
print(hasattr(my_dog, "bark"))     # True (Dogs can bark)
print(hasattr(my_dog, "fly"))      # False (Dogs cannot fly)

# Real world use-case: Safely using a method only if it exists
if hasattr(my_dog, "bark"):
    print(my_dog.bark())
```

## Class Methods & Static Methods
The Three Types of Methods:
1.	**Instance Methods (self):** The standard method. Knows about the specific object's data.
2.	**Class Methods (cls):** Marked with `@classmethod`. Instead of receiving the object (self), it receives the Class itself (cls). Use this when you need to access data that spans across the entire class (like counting how many total objects have been made).
3.	**Static Methods:** Marked with `@staticmethod`. It doesn't receive self or cls. It’s just a regular function that lives inside the class because it logically belongs there (like a utility helper function).

```python
class Pizza:
    # A class-level variable (shared by all pizzas)
    total_pizzas_made = 0

    def __init__(self, toppings):
        self.toppings = toppings
        # Every time a new pizza is built, we increment the global counter
        Pizza.total_pizzas_made += 1

    # 1. Instance Method (Works with a specific pizza's toppings)
    def describe(self):
        print(f"🍕 A delicious pizza with: {', '.join(self.toppings)}")

    # 2. Class Method (Works with the whole class data, uses 'cls' instead of 'self')
    @classmethod
    def get_factory_report(cls):
        print(f"📊 Global Factory Report: Total pizzas made so far = {cls.total_pizzas_made}")

    # 3. Static Method (A helper tool that doesn't need class or instance data)
    @staticmethod
    def convert_inches_to_cm(inches):
        return inches * 2.54


# --- Testing the Different Methods ---

# We can call Class and Static methods directly on the Class itself without making an object!
Pizza.get_factory_report()
print(f"📏 A 12-inch pizza is {Pizza.convert_inches_to_cm(12)} cm.")

print("\n--- Making Pizzas ---")
my_pizza1 = Pizza(["cheese", "pepperoni"])
my_pizza2 = Pizza(["mushrooms", "olives", "onions"])

# Call instance method (needs specific object data)
my_pizza1.describe()

# Call the class method again to see the global updates
Pizza.get_factory_report()
```
