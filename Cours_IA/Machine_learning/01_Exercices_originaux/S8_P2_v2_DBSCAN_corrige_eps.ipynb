```python
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
import pandas as pd
import plotly.express as px
```

```python
X_moons, _ = make_moons(500, noise=0.1, random_state=32)
```

```python
px.scatter(pd.DataFrame({"x": X_moons[:, 0], "y": X_moons[:, 1]}), x="x", y="y", title="Moon Dataset")
```

```python
kmeans = KMeans(n_clusters=2, random_state=32).fit(X_moons)
```

```python
moons = pd.DataFrame(X_moons)
```

```python
moons
```

```python
kmeans.labels_
```

```python
results_df = pd.DataFrame({"x": X_moons[:, 0], "y": X_moons[:, 1], "cluster": kmeans.labels_})
px.scatter(results_df, x="x", y="y", color="cluster", title="K-Means Clustering Results")
```

```python
kmeans.cluster_centers_
```

```python
dbscan = DBSCAN(eps=0.16, min_samples=5).fit(X_moons)
labels_DB = dbscan.labels_
labels_df = pd.concat(
    [results_df.reset_index(drop=True), pd.Series(labels_DB, name="DBSCAN")],
    axis=1
)
labels_df.head()
```

```python
# Visualisation des clusters DBSCAN (le label -1 = bruit / points isolés)
px.scatter(labels_df, x="x", y="y",
           color=labels_df["DBSCAN"].astype(str),
           title="DBSCAN Clustering Results")
```
