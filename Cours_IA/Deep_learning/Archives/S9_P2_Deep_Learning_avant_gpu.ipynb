## Neural Networks


```python
class TinyNetwork (nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(in_features=1, out_features=4)
        self.actication = nn.ReLU()
        self.layer2 = nn.Linear(in_features=4, out_features=1)
    def forward(selt,x) :
        x = selt.layer1(x)
        x = selt.actication(x)
        x = selt.layer2(x)
        return x    
    

```

```python
model = TinyNetwork()
example_input = torch.tensor([[2.0]])
example_output = model(example_input)
print(example_output)
```

```python
x = torch.tensor([[0.0],[1.0],[2.0],[3.0]])
y = torch.tensor([[1.0],[2.0],[4.0],[6.0]])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class TinyDataset(Dataset):
    def __init__(self):
        self.x = torch.tensor([[0.0], [1.0], [2.0], [3.0]])
        self.y = torch.tensor([[1.0], [2.0], [4.0], [6.0]])

    def __len__(self):
        return len(self.x)

    def __getitem__(self, index):
        return self.x[index], self.y[index], 1

dataset = TinyDataset()
print(len(dataset))

x, y, z = dataset[3]
print(f"x:{x} - y:{y}")

loader = DataLoader(dataset, batch_size=2, shuffle=False)

for batch_x, batch_y, batch_z in loader:
    print(f"batch_x = {batch_x}")
```

```python
pure_model = TinyNetwork()
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(pure_model.parameters(), lr=0.01)

for epoch in range(3):
    for i, (batch_x, batch_y, _) in enumerate(loader):
        prediction = pure_model(batch_x)
        loss = loss_fn(prediction, batch_y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        print(f"epoch {epoch} - batch {i} - loss = {loss.item()}")
```
