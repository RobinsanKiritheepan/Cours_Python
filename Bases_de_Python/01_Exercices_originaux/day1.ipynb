## Arithmetic Operators

These are:
- `+`: Sum of two variables.
- `-`: Subtraction of two variables.
- `*`: Product of two variables.
- `/`: Division of two variables.
- `%`: Returns the modulus of two variables, this is the remainer of a division.
- `**`: Exponentiation of two variables.
- `//`: Floor division, i.e., integer division.

```python
a = 3           # a = 3
a = 3 + 2       # a = 5
print("a = ",a)
b = 2           # b = 2
b = a + b       # b = 7
print("b = ",b)
c = a ** b      # c = ?
print("c = ",c)
d = 5.3 // 2    # d = ?
print("d = ",d)
e = b * 5       # e = 35
print("e = ",e)
f = int(5.6)    # f = 5
print("f = ",f)
g = 9 / 3 * (1 + 2)  # 9 or 1 ?
print("g = ",g)
```

## Comparison Operators
Comparison operators take two variables and return a boolean.
- `<`: Smaller than.
- `>`: Greater than.
- `<=`: Smaller or equal than.
- `>=`: Greater or equal than.
- `==`: Equal to.
- `!=`: Unequal to.

## Logical Operators
Logical operators take two boolean variables and return a boolean.
- `and`
- `or`
- `not`

```python
maria_name = "maria"
maria_age = 5
maria_money = 1500.0
iphone_price = 899.9

is_maria_maria = maria_name == 'maria'

is_maria_5 = maria_age == 5

can_maria_drive = maria_age >= 18

can_maria_buy_iphone = maria_money >= iphone_price

# Equivalently
can_maria_buy_iphone = (maria_money - iphone_price) >= 0

# Equivalently
can_maria_buy_iphone = not((maria_money - iphone_price) < 0)

can_maria_drive_and_buy_iphone = can_maria_drive and can_maria_buy_iphone
```

### Dealing with Exceptions
Sometimes Exceptions must not stop code from running...
We use `try` and `except`.

```python
a = 3.0

# We assume b has a value unknown by the programmer, and thus it can be 0
b = input("Input a value for 'b', can be zero.")

try:
    c = a / float(b)
except ZeroDivisionError:
    c = 0
except:
    c = -1

print(c)
```

### Open files and Sys Library

```python
file = open("mon.txt")
print(next(file))
file.close
```

```python

# With context manager, open and close is handled automatically


# Opens as read only
with open("mon.txt") as file:
    print(next(file))


# Opens with writting permission, removes old content
with open("mon.txt", "w") as file:
    file.write("Hello mom\n")


# Opens with writting permission, with old content
with open("mon.txt", "a") as file:
    file.write("Hello dad")
```
