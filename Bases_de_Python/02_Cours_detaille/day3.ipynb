## 1. Les DICTIONNAIRES

Un dictionnaire stocke des paires **clé → valeur**, comme un vrai dictionnaire : tu cherches un MOT (la clé) pour trouver sa DÉFINITION (la valeur).

```python
d = {"nom": "Apollo", "equipage": 3}    # accolades {} et deux-points :
d["nom"]          # → "Apollo"  (accès par clé, pas par position !)
```

**Les 3 règles :**
1. Les **clés** sont UNIQUES et **immutables** (str, nombres, tuples — pas de listes !)
2. Les **valeurs** peuvent être N'IMPORTE QUOI (listes, autres dicts…)
3. Accès, ajout et modification utilisent la même syntaxe : `d[clé] = valeur`

**Liste vs dictionnaire :** une liste s'indexe par POSITION (0, 1, 2…), un dict par CLÉ nommée. Dès que tes données ont des « champs » nommés (nom, âge, prix…), le dict est le bon choix.

### 🔍 Démo vaisseau spatial — 4 opérations fondamentales
1. **Accès** : `spaceship['name']` — et l'enchaînement `spaceship['cargo'][0]` (la valeur est une liste → on l'indexe dans la foulée !)
2. **Modification / ajout** : même syntaxe `d[clé] = valeur` — si la clé existe on écrase, sinon on crée (`destination` apparaît).
3. **Test d'existence** : `if "shields_active" in spaceship:` — sur un dict, `in` teste les CLÉS. Ici → absent… car la vraie clé est `shield_active` (sans s !). Le message d'avertissement s'affiche : c'est exactement le genre de faute de frappe que ce test attrape. 👍
4. **Boucle** : `for key, value in spaceship.items():` — LE pattern de parcours à mémoriser.

````python
# Creating a dictionary to represent a spaceship's status
spaceship = {
    "name": "Apollo Python",
    "crew_count": 3,
    "fuel_level": 84.5,
    "cargo": [" freeze-dried ice cream", "oxygen tanks", "space wrenches"],
    "shield_active": True
}

# 1. Accessing data using keys
print(f"🚀 Spaceship Name: {spaceship['name']}")
print(f"🎒 Cargo Item 1: {spaceship['cargo'][0]}")

# 2. Modifying and Adding new data
spaceship["fuel_level"] = 79.2  # Updating an existing key
spaceship["destination"] = "Mars"  # Adding a completely new key-value pair

# 3. Checking if a key exists safely
if "shields_active" in spaceship:
    print(f"Shield Status: {spaceship['shields_active']}")
else:
    print("⚠️ 'shields_active' key not found! Did you misspell it?")

# 4. Looping through a dictionary
print("\n--- Full Spaceship Diagnostic Report ---")
for key, value in spaceship.items():
    print(f"System [{key.upper()}]: {value}")
````

## 2. Les méthodes de dictionnaires indispensables

| Méthode | Rôle | Pourquoi c'est utile |
|---|---|---|
| `.get(clé, défaut)` | lecture SÛRE | `d[clé]` PLANTE (`KeyError`) si la clé manque ; `.get` renvoie le défaut à la place — jamais de crash ! |
| `.pop(clé)` | retire la clé ET renvoie sa valeur | supprimer en récupérant |
| `.keys()` / `.values()` | toutes les clés / valeurs | inventaire rapide |
| `.items()` | les paires (clé, valeur) | LA façon de boucler : `for k, v in d.items():` |
| `.update(autre_dict)` | fusionne un autre dict | les clés communes sont ÉCRASÉES par les nouvelles |

### 🔍 Démo maison connectée — les méthodes en action
1. `.get("backyard_temp", "Sensor Offline")` → la clé n'existe pas → le DÉFAUT est renvoyé, pas de crash. Compare : `smart_home["backyard_temp"]` aurait levé `KeyError` !
2. `list(d.keys())` / `list(d.values())` → inventaires (le `list()` rend l'affichage propre).
3. `.pop("eco_mode")` → retire la clé ET récupère sa valeur (False) au passage.
4. `.update(night_preset)` → fusion : `living_room_temp` et `lights_on` sont ÉCRASÉS par les valeurs du preset, `camera_recording` (nouvelle clé) est AJOUTÉE.

````python
# A dictionary representing a smart home system configuration
smart_home = {
    "living_room_temp": 22,
    "lights_on": ["kitchen", "hallway"],
    "security_armed": True,
    "eco_mode": False
}

# 1. Safe Access with .get()
# Standard bracket lookups crash if the key is missing. .get() handles it gracefully.
thermostat = smart_home.get("backyard_temp", "Sensor Offline")
print(f"🌡️ Backyard Temperature: {thermostat}")

# 2. Extracting Keys and Values independently
print(f"🔑 Available Settings: {list(smart_home.keys())}")
print(f"📊 Current Values: {list(smart_home.values())}")

# 3. Removing items safely using .pop()
# Let's turn off eco mode and capture its last state
was_eco = smart_home.pop("eco_mode")
print(f"♻️ Removed 'eco_mode'. Its last setting was: {was_eco}")

# 4. Merging dictionaries with .update()
# Imagine the user applies a 'Night Mode' preset
night_preset = {
    "living_room_temp": 19,
    "lights_on": ["porch"],
    "camera_recording": True # This is a brand new setting!
}

smart_home.update(night_preset)

print("\n--- Updated Smart Home Status (Night Mode) ---")
for setting, status in smart_home.items():
    print(f"{setting}: {status}")
````

## 3. `*args` : un nombre VARIABLE d'arguments positionnels

Comment écrire une fonction qui accepte 2, 5 ou 100 arguments (comme `print` !) ?

```python
def f(param_normal, *args):   # l'étoile fait la magie
```
À l'appel, Python **emballe** tous les arguments supplémentaires dans un **TUPLE** nommé `args`. Dans la fonction, tu le traites comme un tuple normal : `len(args)`, boucle `for`… (Le nom `args` est une convention — c'est l'étoile `*` qui compte.)

### 🔍 Démo `*args` — regarde les preuves dans la sortie :
- `type(cities)` → `<class 'tuple'>` : les 3 villes ("Tokyo", "Kyoto", "Osaka") ont été EMBALLÉES automatiquement.
- Le 1er paramètre (`desination_country` — coquille pour *destination* 😉) reçoit "Japan" normalement ; SEULS les arguments en surplus vont dans `*cities`.
- La fonction gère aussi le cas zéro ville (`len(cities) == 0`) — bonne pratique.

````python
def plan_trip(desination_country: str, *cities):
    print(f"Trip planner to {desination_country.upper()}")
    print(f"The type of 'cities' is {type(cities)}")
    print(f"Content of 'cities': {cities}")

    # Because 'cities' is a tuple, we can check its length and loop through it!
    if len(cities) == 0:
        print("Warning: No cities added to this itinerary yet.")
    else:
        print("Stops along the way:")
        for city in cities:
            print(f"   {city}")

plan_trip("Japan", "Tokyo", "Kyoto", "Osaka")
````

### 📝 EXERCICE — `multiply_all(*numbers)`

**Énoncé (traduit) :** écrire une fonction qui accepte un nombre quelconque d'arguments numériques et renvoie leur PRODUIT. Cas particulier : aucun argument → renvoyer 0 (selon l'énoncé).

**Solution commentée :**
- `product = 1` puis `product *= n` pour chacun — 1 est l'élément NEUTRE de la multiplication (démarrer à 0 donnerait toujours 0 !).
- Le cas vide renvoie 0 comme demandé par l'énoncé. 💡 Anecdote mathématique : le « produit vide » vaut conventionnellement 1 (comme la somme vide vaut 0) — mais on suit l'énoncé.
- Tests : 2×3×4 = 24 ✓, 5×10×(−5)×9 = −2250 ✓ (le signe survit !), () → 0 ✓.

````python
def multiply_all(*numbers):
    if len(numbers) == 0:
        return 0
    product = 1
    for n in numbers:
        product *= n
    return product
    
    
    
print(multiply_all(2, 3, 4))        # Expected Output: 24
print(multiply_all(5, 10, -5, 9))   # Expected Output: -2250
print(multiply_all())               # Expected Output: 0
````

## 4. `**kwargs` : un nombre variable d'arguments NOMMÉS

La double étoile emballe les arguments **nommés** supplémentaires… dans un **DICTIONNAIRE** !

```python
def f(nom, **kwargs): ...
f("Albert", ville="Princeton", prix_nobel=True)
# → kwargs = {"ville": "Princeton", "prix_nobel": True}
```
**Le duo récapitulé :** `*args` → tuple d'arguments positionnels en plus ; `**kwargs` → dict d'arguments nommés en plus. Ensemble, ils permettent des fonctions totalement flexibles (tu les reverras dans les décorateurs du Jour 4 !).

### 🔍 Démo `**kwargs` — les preuves :
- `type(additional_info)` → `<class 'dict'>` !
- Scénario A (aucun extra) → `kwargs = {}` (dict vide, la fonction marche quand même).
- Scénario B → `{"location": "Princeton", "field": "Physics", "nobel_prize": True}` — les 3 arguments nommés emballés.
- La boucle `.items()` recopie ensuite ces extras dans le profil : le dict de sortie fusionne le fixe et le variable.

````python
def build_profile(first_name, last_name, **additional_info):
    """Builds a dictionary containing everything we know about a user."""
    profile = {
        "first": first_name,
        "last": last_name,
    }
    
    # additional_info is a dictionary containing any extra keyword arguments passed!
    print(f"The type of additional_info is {type(additional_info)}")
    print(f"Content of **kwargs: {additional_info}\n")
    
    for key, value in additional_info.items():
        profile[key] = value
        
    return profile

# Scenario A: Minimal info
user1 = build_profile("Ada", "Lovelace")
print("User 1 Profile:", user1)
print("-" * 20)

# Scenario B: Lots of optional info
user2 = build_profile("Albert", "Einstein", 
                      location="Princeton", 
                      field="Physics", 
                      nobel_prize=True)
print("User 2 Profile:", user2)
````

## 5. La Programmation Orientée Objet (POO)

Jusqu'ici : programmation **procédurale** (des fonctions qui s'enchaînent). La POO regroupe **données + comportements** dans un même moule :

| Concept | Définition | Analogie |
|---|---|---|
| **Classe** | le moule / plan de construction | le plan d'architecte d'une maison |
| **Objet / Instance** | un exemplaire concret créé depuis le moule | LA maison au 123 rue Python |
| **Attribut** | une variable DANS l'objet | la couleur de la maison |
| **Méthode** | une fonction DANS l'objet | ouvrir_porte() |

**La syntaxe à décoder (cellule suivante) :**
- `class VirtualPet:` → on définit le moule.
- `def __init__(self, name, species):` → le **constructeur** : exécuté automatiquement à CHAQUE création d'objet ; il initialise les attributs.
- **`self`** = « l'objet lui-même » — le paramètre obligatoire de toute méthode ; `self.name` = « l'attribut name de CET objet ». Quand tu appelles `pet1.play(20)`, Python passe `pet1` comme `self` tout seul.

### 🔍 Démo VirtualPet — suis le cycle de vie :
1. `pet1 = VirtualPet("Sparky", "Dog")` → Python exécute `__init__` : self.name="Sparky", self.energy=100. `pet2` pareil avec SES valeurs — **deux objets indépendants issus du même moule**.
2. `pet1.play(20)` → la méthode modifie `self.energy` : 100 − 20×2 = 60. Vérifie dans la sortie : le 2e check_status de Sparky affiche 60 ✅ — l'objet a une MÉMOIRE interne.
3. `pet2.play(45)` → 100 − 90 = 10 → le message « exhausted » se déclenche (energy < 20). Et l'énergie de pet1 n'a pas bougé : chaque objet vit sa vie !

````python
class VirtualPet:
    # The Constructor method - runs automatically when a new pet is created
    def __init__(self, name, species):
        self.name = name          # Attribute
        self.species = species    # Attribute
        self.energy = 100         # Default Attribute value

    # A Method representing behavior
    def play(self, minutes):
        energy_lost = minutes * 2
        self.energy -= energy_lost
        print(f"🎮 {self.name} played for {minutes} minutes and lost {energy_lost} energy.")
        if self.energy < 20:
            print(f"😴 {self.name} is exhausted! Time for a nap.")

    # A Method to check status
    def check_status(self):
        print(f"📊 Pet Status -> Name: {self.name} | Species: {self.species} | Energy: {self.energy}%")


# --- Putting the Blueprint to Work (Instantiating Objects) ---

# Creating two distinct pets from the same blueprint
pet1 = VirtualPet("Sparky", "Dog")
pet2 = VirtualPet("Whiskers", "Cat")

# Interacting with our objects
pet1.check_status()
pet1.play(20)
pet1.check_status()

print("-" * 30)

pet2.check_status()
pet2.play(45)
````

## 6. L'HÉRITAGE

**Le problème :** Warrior, Mage et Rogue ont tous un nom, des PV, une méthode take_damage()… Copier-coller le code 3 fois viole le principe **DRY** (*Don't Repeat Yourself* — ne te répète pas !).

**La solution :** une classe PARENTE générique (`Media`) dont HÉRITENT des classes ENFANTS (`Movie`, `Series`) — elles reçoivent GRATUITEMENT tous les attributs et méthodes du parent, et peuvent :
- **ajouter** leurs spécificités (`director` pour Movie) ;
- **redéfinir** (*override*) un comportement hérité (`Series.play()` remplace `Media.play()`).

**Les deux syntaxes clés :**
- `class Movie(Media):` → « Movie hérite de Media »
- `super().__init__(...)` → « appelle le constructeur du PARENT » (pour ne pas réécrire l'initialisation commune)

**Bonus — `__str__` :** cette méthode spéciale définit ce que `print(objet)` affiche. Sans elle : `<Media object at 0x...>` (illisible) ; avec elle : un joli texte de ton choix !

### 🔍 Démo héritage — les 3 points à vérifier dans la sortie :
1. **`__str__` en action :** `print(generic_video)` affiche le texte défini par Media ; `print(movie1)` affiche la VERSION REDÉFINIE par Movie (avec le réalisateur) — chaque classe son style.
2. **Héritage silencieux :** `movie1.play()` → Movie n'a PAS de méthode play… Python remonte au parent Media et utilise la sienne ! (« ▶️ Playing… »)
3. **Redéfinition :** `series1.play()` → Series A sa propre play → « 📺 Binging… ». Même appel, comportements différents selon la classe : c'est le **polymorphisme**.
Et les deux compteurs de vues finissent à 1 chacun — l'attribut `views` (hérité de Media) vit indépendamment dans chaque objet.

````python
# The Parent Class (Generic Blueprint)
class Media:
    def __init__(self, title, duration_mins):
        self.title = title
        self.duration_mins = duration_mins
        self.views = 0

    def play(self):
        self.views += 1
        print(f"▶️ Playing: '{self.title}'")

    # Special method to handle how this object prints as a string
    def __str__(self):
        return f"'{self.title}' ({self.duration_mins} mins) - Views: {self.views}"


# The Child Class (Inherits everything from Media)
class Movie(Media):
    # Overriding the parent constructor to add a specific attribute
    def __init__(self, title, duration_mins, director):
        # super().__init__ calls the parent class's constructor to set title and duration
        super().__init__(title, duration_mins)
        self.director = director  # Specific to Movies

    # Overriding the parent's print string representation
    def __str__(self):
        return f"🎬 Movie: '{self.title}' directed by {self.director}"


# Another Child Class
class Series(Media):
    def __init__(self, title, duration_mins, total_episodes):
        super().__init__(title, duration_mins)
        self.total_episodes = total_episodes

    # Overriding the play method specifically for Series behavior
    def play(self):
        self.views += 1
        print(f"📺 Binging an episode of '{self.title}' ({self.total_episodes} eps total)")


# --- Testing Inheritance & Special Methods ---

# 1. Look how clean the print statements look now because of __str__
generic_video = Media("Funny Cat Compilation", 5)
print(generic_video) 

movie1 = Movie("Inception", 148, "Christopher Nolan")
print(movie1) 

series1 = Series("Python for Beginners Show", 45, 10)
print(series1)

print("-" * 40)

# 2. Testing inherited and overridden methods
movie1.play()       # Uses parent's play() method logic
series1.play()      # Uses its own overridden play() method logic

print(f"Movie Views: {movie1.views}")
print(f"Series Views: {series1.views}")
````

## 7. Inspecter les objets

| Fonction | Question posée |
|---|---|
| `isinstance(obj, Classe)` | « cet objet est-il de cette classe (ou d'une classe enfant) ? » |
| `issubclass(Enfant, Parent)` | « cette CLASSE hérite-t-elle de celle-là ? » |
| `hasattr(obj, "nom")` | « cet objet possède-t-il cet attribut/méthode ? » |

**Subtilité importante démontrée ci-dessous :** un objet `Dog` est AUSSI une instance d'`Animal` (`isinstance(my_dog, Animal)` → True) — l'héritage crée une relation « EST UN ». Usage pro de `hasattr` : vérifier qu'une méthode existe avant de l'appeler.

*(⚠️ Petite erreur dans le support de cours original : `issuperclass` n'existe pas en Python ! Seul `issubclass` existe — pour tester dans l'autre sens, on inverse simplement les arguments.)*

### 🔍 Démo inspection — les résultats clés :
- `issubclass(Dog, Animal)` → True, l'inverse → False (l'héritage a un sens unique !)
- `isinstance(my_dog, Animal)` → **True** alors que my_dog est un Dog — un chien EST un animal, l'héritage se propage.
- `hasattr(my_dog, "bark")` → True / `"fly"` → False, puis l'usage défensif : n'appeler `.bark()` QUE s'il existe.

````python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def bark(self):
        return "Woof!"

# Let's create a dog object
my_dog = Dog("Buddy")

# 1. Checking Class Relationships with issubclass()
print("--- Class Checks ---")
print(issubclass(Dog, Animal))     # True (Dog is a child of Animal)
print(issubclass(Animal, Dog))     # False (Animal is not a child of Dog)

# 2. Checking Objects with isinstance()
print("\n--- Object Checks ---")
print(isinstance(my_dog, Dog))     # True (It is a Dog)
print(isinstance(my_dog, Animal))  # True (It is also an Animal because of inheritance!)

# 3. Checking Features with hasattr()
print("\n--- Feature Checks ---")
print(hasattr(my_dog, "bark"))     # True (Dogs can bark)
print(hasattr(my_dog, "fly"))      # False (Dogs cannot fly)

# Real world use-case: Safely using a method only if it exists
if hasattr(my_dog, "bark"):
    print(my_dog.bark())
````

## 8. Les trois types de méthodes

| Type | Décorateur | Reçoit | Usage |
|---|---|---|---|
| **d'instance** | (aucun) | `self` (l'objet) | le cas normal : travailler avec LES données de CET objet |
| **de classe** | `@classmethod` | `cls` (la classe) | travailler avec des données PARTAGÉES par tous les objets (ex : un compteur global d'instances) |
| **statique** | `@staticmethod` | rien | une fonction utilitaire qui vit dans la classe juste parce qu'elle y a sa place logique |

**Dans la démo Pizza :** `total_pizzas_made` est une variable DE CLASSE (partagée !) — chaque `__init__` l'incrémente via `Pizza.total_pizzas_made`. `get_factory_report` (classmethod) la lit. `convert_inches_to_cm` (staticmethod) est juste un convertisseur. Remarque : on peut appeler les deux dernières SANS créer d'objet : `Pizza.get_factory_report()` !

### 🔍 Démo Pizza — suis le compteur partagé :
1. AVANT toute création : `Pizza.get_factory_report()` → 0 pizzas (appel sur la CLASSE, sans objet !), et le convertisseur statique → 30.48 cm.
2. On crée 2 pizzas → chaque `__init__` fait `Pizza.total_pizzas_made += 1` sur la variable DE CLASSE (partagée).
3. `my_pizza1.describe()` → méthode d'instance : elle parle des toppings de CETTE pizza.
4. Le rapport final → **2** : le compteur a bien enregistré les deux créations, tous objets confondus.

## 📝 Résumé du Jour 3

4. **Héritage** : `class Enfant(Parent)`, `super().__init__()`, redéfinition de méthodes, `__str__` pour un print lisible.
5. Trois méthodes : instance (`self`), classe (`cls`, données partagées), statique (utilitaire).

````python
class Pizza:
    # A class-level variable (shared by all pizzas)
    total_pizzas_made = 0

    def __init__(self, toppings):
        self.toppings = toppings
        # Every time a new pizza is built, we increment the global counter
        Pizza.total_pizzas_made += 1

    # 1. Instance Method (Works with a specific pizza's toppings)
    def describe(self):
        print(f"🍕 A delicious pizza with: {', '.join(self.toppings)}")

    # 2. Class Method (Works with the whole class data, uses 'cls' instead of 'self')
    @classmethod
    def get_factory_report(cls):
        print(f"📊 Global Factory Report: Total pizzas made so far = {cls.total_pizzas_made}")

    # 3. Static Method (A helper tool that doesn't need class or instance data)
    @staticmethod
    def convert_inches_to_cm(inches):
        return inches * 2.54


# --- Testing the Different Methods ---

# We can call Class and Static methods directly on the Class itself without making an object!
Pizza.get_factory_report()
print(f"📏 A 12-inch pizza is {Pizza.convert_inches_to_cm(12)} cm.")

print("\n--- Making Pizzas ---")
my_pizza1 = Pizza(["cheese", "pepperoni"])
my_pizza2 = Pizza(["mushrooms", "olives", "onions"])

# Call instance method (needs specific object data)
my_pizza1.describe()

# Call the class method again to see the global updates
Pizza.get_factory_report()
````
